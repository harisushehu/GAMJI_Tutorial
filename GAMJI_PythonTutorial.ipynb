```python
#read data
df = pd.read_csv("/content/2019elections.csv")
df.head()
```

```python
# Calculate the average percentage of valid votes
avg_valid_votes = df['Valid votes'].mean() / registered_voters * 100
print("Average valid votes percentage", avg_valid_votes)

# Find the state with the highest PDP votes
max_pdp_votes_state = df.loc[df['PDP Votes'].idxmax()]['State']
print("State with Highest PDP Votes", max_pdp_votes_state)

# Calculate the average APC percentage votes across all states
avg_apd_percent = df['APC % Votes'].mean()
print("Average % APC votes", avg_apd_percent)


# Calculate the total number of votes cast in the South South geopolitical zone
south_south_votes = df[df['Geopolitical zone']== 'South South']['Votes cast'].sum()
print("Total votes cast in SS: ", south_south_votes)
```

```python
#plot bar chart depicting the number of votes by states

import numpy as np
import matplotlib.pyplot as plt

states = df['State'][0:5]
values = df['Votes cast'][0:5]

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(states, values, color ='maroon',
		width = 0.4)

plt.xlabel("State")
plt.ylabel("Votes cast")
plt.title("Number of votes by states")
plt.show()
```

```python
#plot bar chart depicting the number of APC and PDP votes by states

fig = plt.subplots(figsize = (12, 8))

data_1 = df['APC Votes'][0:5]
data_2 = df['PDP Votes'][0:5]

barWidth = 0.25
bar1 = np.arange(len(data_1))
bar2 = [x + barWidth for x in bar1]

plt.bar(bar1, data_1, color = 'r', width = barWidth, edgecolor = 'grey', label = 'APC')
plt.bar(bar2, data_2, color = 'g', width = barWidth, edgecolor = 'grey', label = 'PDP')

plt.xticks([r + (0.5 * barWidth) for r in range(len(data_1))], ['Abia', 'Adamawa', 'Akwai Ikom', 'Anambra', 'Bauchi'])

plt.xlabel('States', fontweight = 'bold', fontsize = 15)
plt.ylabel('Number of Votes', fontweight = 'bold', fontsize = 15)


plt.legend(title = 'Party', loc = 'upper left')

plt.show()
```

```python
# Calculate the sum of 'Votes cast' grouped by 'Geopolitical zone'
pie_data = df["Votes cast"].groupby(df['Geopolitical zone']).sum()
pie_data = pie_data.values
#print(pie_data)

pie_label = df['Geopolitical zone'].unique()
#print(pie_label)

#plot pie chart depicting the distribution of votes by geopolitical zones
import seaborn

palette_color = seaborn.color_palette('bright') #dark

plt.pie(pie_data, labels = pie_label, colors = palette_color, autopct = '%0.0f%%')
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import plotly.express as px

# Read the JSON shape file
nigeria_map = gpd.read_file("nigeria_regions.json")  # Replace with the actual path


data.loc[data['State'] == 'FCT', 'State'] = "Federal Capital Territory"

# Merge the election data with the map data based on state names
merged = nigeria_map.merge(data, left_on="ID", right_on="State")

# Create a choropleth map
fig = px.choropleth(merged,
                    geojson=merged.geometry,
                    locations=merged.index,
                    color="PDP Votes",
                    hover_name="State",
                    hover_data=["APC Votes", "PDP Votes"],  # Include additional hover data
                    color_continuous_scale="Viridis",
                    title="Choropleth Map of Votes Cast by State in Nigeria")

# Customize hover template to include APC and PDP votes
fig.update_traces(hovertemplate="<b>%{hovertext}</b><br>APC Votes: %{customdata[0]:,}<br>PDP Votes: %{customdata[1]:,}")

# Hide the axis and show the map
fig.update_geos(showcoastlines=True, coastlinecolor="Black", showland=True, landcolor="white")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
